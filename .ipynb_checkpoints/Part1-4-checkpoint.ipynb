```python
import pandas as pd
from scipy.stats import iqr
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("part1-3.csv")
```

```python
#데이터 전처리
is_EB_data = data['Age'] == "Early Bronze Age"
EB_data = data[is_EB_data]

is_LB_data = data['Age'] == "Late Bronze Age"
LB_data = data[is_LB_data]
```

```python
#문제 1
fig, ax = plt.subplots()
ax.boxplot([EB_data['Site(ha)'], LB_data['Site(ha)']], sym='bo')
plt.title("Size of Nanxiong Site")
plt.xticks([1, 2], ["Early Bronze Age", "Late Bronze Age"])
plt.grid(True)
plt.show()
```

```python
#데이터 전처리 : 수준 제거
EB_median = EB_data["Site(ha)"].median()
LB_median = LB_data["Site(ha)"].median()

print("Early Bronze Age median : %0.1f" % EB_median)
print("Late Bronze Age median : %0.1f" % LB_median)

EB_remove_levels = EB_data['Site(ha)'] - EB_median
LB_remove_levels = LB_data['Site(ha)'] - LB_median
```

```python
#문제 2
fig, ax = plt.subplots()
ax.boxplot([EB_remove_levels, LB_remove_levels], sym='bo')
plt.title("Size of Nanxiong Site(Remove Levels)")
plt.xticks([1, 2], ["Early Bronze Age", "Late Bronze Age"])
plt.grid(True)
plt.show()
```

```python
#데이터 전처리 : 산포 제거
EB_iqr = iqr(EB_data["Site(ha)"])
LB_iqr = iqr(LB_data["Site(ha)"])

EB_remove_levels_iqr = EB_remove_levels / EB_iqr
LB_remove_levels_iqr = LB_remove_levels / EB_iqr
```

```python
#문제 3
fig, ax = plt.subplots()
ax.boxplot([EB_remove_levels_iqr, LB_remove_levels_iqr], sym='bo')
plt.title("Size of Nanxiong Site(Remove Levels&IQR)")
plt.xticks([1, 2], ["Early Bronze Age", "Late Bronze Age"])
plt.grid(True)
plt.show()
```
