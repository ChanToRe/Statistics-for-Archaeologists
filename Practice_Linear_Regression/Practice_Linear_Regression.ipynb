```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
train_data = pd.read_csv("LR_train_data.csv") #학습 데이터
```

```python
is_Paleolithic_data = train_data['Period'] == '구석기' #학습 데이터 전처리
Paleolithic_data = train_data[is_Paleolithic_data]
is_ER_data = Paleolithic_data['E.R'] <= 150
ER_data = Paleolithic_data[is_ER_data]
```

```python
plt.scatter(ER_data['BP'], ER_data['Latitude']) #
plt.title("ABCD Country")
plt.xlabel("BP")
plt.ylabel("Latitude")
plt.axis([19000, 45000, 34, 43])
plt.show() 
```

```python
X = ER_data["BP"]
Y = ER_data["Latitude"]
LM = LinearRegression()
LM.fit(X.values.reshape(-1,1), Y)
```

```python
LM.predict([[25000]])
```

```python
plt.plot(X, LM.predict(X.values.reshape(-1,1)))
plt.scatter(ER_data['BP'], ER_data['Latitude']) 
plt.title("ABCD Country")
plt.xlabel("BP")
plt.ylabel("Latitude")
plt.axis([45000, 19000, 34, 43])
```

```python
print("회귀계수 a : %f" % LM.coef_)
print("상수 b : %f" % LM.intercept_)
```

```python
print("Socre : %f" % LM.score(X.values.reshape(-1,1), Y))
```
