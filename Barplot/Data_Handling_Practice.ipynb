```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
data = pd.read_csv('rcdb.csv')
plt.rcParams["figure.figsize"] = (35,5)
```

```python
#matplotlib 사용
plot = data['State'].value_counts().reset_index()

plt.bar(plot['index'], plot['State'])
plt.title("광역시/도/특별시 별 탄소연대측정 수", fontsize=25)
plt.xlabel("광역시/도/특별시", fontsize=20)
plt.ylabel("측정수", fontsize=20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid()
plt.show()
```

```python
#Pandas 내장 기능 사용
data["State"].value_counts().plot(kind='bar')
plt.title("광역시/도/특별시 별 탄소연대측정 수", fontsize=25)
plt.xlabel("광역시/도/특별시", fontsize=20)
plt.ylabel("측정수", fontsize=20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid()
plt.show()
```

```python
data_dup = data.drop_duplicates(['Report'])
data_dup["State"].value_counts().plot(kind='bar')
plt.title("광역시/도/특별시 별 탄소연대측정수(중복제거)", fontsize=25)
plt.xlabel("광역시/도/특별시", fontsize=20)
plt.ylabel("측정수", fontsize=20)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.grid()
plt.show()
```
